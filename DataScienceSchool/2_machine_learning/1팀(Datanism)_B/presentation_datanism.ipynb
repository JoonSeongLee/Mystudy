     * Cosine similary :
     $\frac{
\sum\limits_{i \in I_{uv}} r_{ui} \cdot r_{vi}}
{\sqrt{\sum\limits_{i \in I_{uv}} r_{ui}^2} \cdot
\sqrt{\sum\limits_{i \in I_{uv}} r_{vi}^2}
}
$



- 사용한 Feature : 
    - CAPSULE_TEXT
    - DISCOUNT_PRICE
    - large_area_name

#### 4.1. 판매된 쿠폰에 대한 Feature들을 나타낸 Table 만들기 

 = train 데이터, 아웃라이어 제거 후 168996개에서 142474개의 판매기록

````python
purchased_coupons_train = coupon_purchases_train.merge(coupon_list_train, on='COUPON_ID_hash', how='inner')
purchased_coupons_train = purchased_coupons_train.merge(user_list, on = 'USER_ID_hash', how = 'outer')
features = ['COUPON_ID_hash', 'USER_ID_hash', 'CAPSULE_TEXT', 'DISCOUNT_PRICE','large_area_name', 'ken_name', 'small_area_name',]
purchased_coupons_train = purchased_coupons_train[features]
print('판매기록 수 : ',len(purchased_coupons_train))
purchased_coupons_train.head(10) 
````

#### 4.2. Category Feature들 인코딩

````python
# encoding categorical columns 
coupon_list_test['USER_ID_hash'] = 'tmp'
coupon_list_test = coupon_list_test[features]
combined = pd.concat([purchased_coupons_train, coupon_list_test], axis=0)
categoricals = ['CAPSULE_TEXT', 'large_area_name', 'ken_name', 'small_area_name',]
combined_categoricals = combined[categoricals]
combined_categoricals = pd.get_dummies(combined_categoricals, dummy_na=False)
# concat numeric and category
continuous = list(set(features) - set(categoricals))
combined = pd.concat([combined[continuous], combined_categoricals], axis=1)
combined.head(10)
````

#### 4.3. NaN값 처리 및 test 파일 분리

````python
### remove NaN values
NAN_SUBSTITUTION_VALUE = 1
combined = combined.fillna(NAN_SUBSTITUTION_VALUE)
# split test data
train = combined[combined['USER_ID_hash'] != 'tmp']
test = combined[combined['USER_ID_hash'] == 'tmp']
test2 = test.copy()
test2.drop('USER_ID_hash', inplace=True, axis=1)
````

#### 4.4. 유사도 비교를 위한 Table 만들기

````python
train_dropped_coupons = train.drop('COUPON_ID_hash', axis=1)
train_fin = train_dropped_coupons.groupby(by='USER_ID_hash').mean()
test_fin = test2.drop('COUPON_ID_hash', axis=1)
````

##### 4.4.1 train data 확인

##### 4.4.2 test data 확인

````python
print('rows : ',len(test_fin))
test_fin.head()
````

#### 4.5. Scaling

````python
vector_1 = train_fin.copy()
vector_2 = test_fin.copy()

vector_1 = scale(vector_1)
vector_1 = pd.DataFrame(data = vector_1, index = train_fin.index, columns = train_fin.columns )

vector_2 = scale(vector_2)
vector_2 = pd.DataFrame(data = vector_2, index = test_fin.index, columns = test_fin.columns )
````

#### 4.6. cosine_similarity

````python
def cosine_similarity(vector_1, vector_2):
    return 1 - spatial.distance.cosine(vector_1, vector_2)
````
