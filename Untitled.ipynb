`회색`

\colorbox{Brown}{text}

$$
(x)\approx\frac{\phi\prime (x)\times(1-p\prime)\times p}
{\phi\prime(x)\times(1-p\prime)\times p +(1-\phi\prime(x)\times 
p\prime\times(1-p)}\\
$$
