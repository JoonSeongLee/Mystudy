## 2.1 분류와 회귀

지도학습은 분류와 회귀가 있습니다.

**분류**는 미리 정의된, 가능성 있는 야러 클레스 레이즐 중 하나를 예측하는 것입니다.

분류는 딱 두 개의 클래스로 분류하는 **이진 분류**(양성 클래스 / 음성 클래스)

셋이상의 클래스로 분류하는 **다중 분류**로 나뉩니다.

**회귀**는 연속적인 숫자, 또는 프로그래밍 용어로 말하면 부동소수점수(실수)를 예측하는 것입니다.


출력 값에 연속성이 있는지 질문해보면 회귀와 분류 문제를 쉽게 구분할 수 있습니다.

## 2.2 일반화, 과대적합, 과소적합

지도학습에서는 훈련 데이터로 학습한 모델이 새로운 데이터를 예측할 거라 기대합니다. 모델이 처음 보는 데이터를 정확하게 예측할 수 있으면 이를 훈련 세트에서 데이터 세트로 **일반화**되었다고 합니다.

너무 복잡한 모델을 만드는 것을 **과대적합(overfitting)**이라고 합니다. 훈련 세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때 일어납니다.

너무 간단한 모델이 선택되는 것을 **과소적합**이라고 합니다.

### 2.2.1 모델 복잡도와 데이터셋 크기의 관계

모델의 복잡도는 훈련 데이터셋에 담긴 입력 데이터의 다양성과 관련이 깊습니다.

다양한 데이터 포인트가 많다 -> 과대적합 없이 더 복잡한 모델을 만들 수 있습니다.

하지만 같은 데이터 포인트를 중복하거나 매우 비슷한 데이터를 모으는 것은 도움이 안 됩니다.
